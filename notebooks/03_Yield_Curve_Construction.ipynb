# Module 3 - Yield Curve Construction

## Overview

A yield curve represents the relationship between interest rates and time to
maturity. It is one of the most important building blocks in quantitative
finance.

Instead of pricing instruments directly from market quotes, pricing engines
first construct a zero-coupon yield curve. The curve provides discount
factors for valuing future cashflows and serves as the foundation for pricing
interest rate derivatives.

In this notebook we will:

- Create market quotes
- Bootstrap a yield curve
- Store curve points
- Query zero rates
- Calculate discount factors
- Interpolate rates for intermediate maturities

## Step 1: Import Required Modules

Import the classes developed in this module.
These classes collectively transform market quotes into a reusable yield curve.

```python
from datetime import date

from src.market_data.market_instrument import MarketInstrument
from src.market_data.market_quote_collection import MarketQuoteCollection
from src.curves.bootstrapper import Bootstrapper
```

## Step 2: Create Market Quotes

Market quotes are the observable interest rates available from the market.

For this example, we use deposit rates with maturities of:

- 1 Month
- 3 Months
- 6 Months
- 1 Year

These quotes will be the inputs to the bootstrapping process.   

```python
quotes = MarketQuoteCollection()

quotes.add(MarketInstrument("Deposit", "1M", 0.048))
quotes.add(MarketInstrument("Deposit", "3M", 0.050))
quotes.add(MarketInstrument("Deposit", "6M", 0.052))
quotes.add(MarketInstrument("Deposit", "1Y", 0.055))

quotes.summary()
```

## Step 3: Bootstrap the Yield Curve

The Bootstrapper converts market quotes into a sequence of zero-coupon
curve points.

Each CurvePoint stores:

- Tenor
- Time to maturity
- Zero rate
- Discount factor

The resulting YieldCurve can later be used by pricing engines.

```python
bootstrapper = Bootstrapper(
    valuation_date=date.today(),
    market_quotes=quotes,
)

curve = bootstrapper.build()
```

## Step 4: Display the Yield Curve

The YieldCurve stores an ordered collection of CurvePoint objects.

Each point contains both the zero rate and its corresponding discount factor,
making the curve compact, immutable, and easy to query.

```python
curve.summary()
```

## Step 5: Query the Yield Curve

Pricing engines frequently require zero rates or discount factors for
specific maturities.

The YieldCurve provides convenient lookup methods using either market
tenors or time to maturity.

```python
print(curve.get_zero_rate("6M"))

print(curve.get_discount_factor("1Y"))
```

## Step 6: Linear Interpolation

Market quotes are only available for a limited set of maturities.

To estimate rates between quoted maturities, we apply linear interpolation.

Although simple, this technique is widely used in educational pricing
libraries and provides a solid foundation before introducing more advanced
curve-fitting methods.

```python
curve.get_zero_rate_by_time(0.75)
```
